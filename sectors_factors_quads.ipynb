```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from datetime import datetime
plt.style.use('fivethirtyeight')
# plt.style.use('seaborn')
%matplotlib inline 
import yfinance as yf
idx = pd.IndexSlice
```

```python
tickers = 'SPY IWM QQQ RSP XLE XLP XLI XLF XLY XLB XLRE XLK XLC XLU XLV IWF IWD VYM USMV MTUM TLT TIP IEF'
start=None
end=None
```

```python
data = yf.download(tickers, start=start, end=end, auto_adjust=True,
                   group_by="ticker")
```

```python
prices = data.loc[:, idx[:,'Close']]
prices.columns = prices.columns.droplevel(1)
```

```python
prices
```

```python
# get quad data 
quads = pd.read_csv('quads_06-1948_09-2020.csv', index_col=0, parse_dates=['Date '])
quads
```

```python
def period_total_return(df):
    period_total_returns = df.add(1).cumprod().sub(1)
    return(period_total_returns[-1])
```

```python
returns = prices.pct_change().resample('bQ').apply(period_total_return)
```

```python
returns_quads = returns.join(quads)
```

```python
returns_quads.loc['1998':].groupby('Quad').mean().mul(100).round(2).T
```
